### (2 -> 12 -> 1)

In Input layer, W1 and b1 exist and have 2*12 and 1*12 order
        act func are sig, tanh and relu
In Hidden layer, W2 and b2 exist and have 12*1 and 1*1 order
        act func as sig
In Output layer, y of order 1*1

```python
def sig(x):
    return 1/(1+np.exp(-x))

class NN:
    def __init__(self, act="sig", lr=0.01):
        self.lr= lr
        self.act= act
        np.random.seed(0)
        self.W1= np.random.randn(2,12)*0.1
        self.b1= np.zeros((1,12))
        self.W2= np.random.randn(12,1)*0.1
        self.b2= np.zeros((1,1))
    
    def forward(self,X):
        self.H1= X @ self.W1 + self.b1
        self.A1= self.act(self.Z1)
        self.H2= self.A1 @ self.W2 + self.b2
        self.A2= sig(self.Z2)
        return self.A2

    def backward(X, W, b):
        # w_new = w - lr * grd(x)
        W= W - self.lr * self.act_d(X)
        b= b - self.lr * self.act_d(X)
        return  W, b

    
    def train(X, y):
        losses= []
        for i in range(10000):
            y_pred= self.forward(X)
            losses= self.loss(y_pred, y)
            print(losses)
            self.backward(X, W, b)
            acc= self.accuracy(y_pred, y)

    def act(self,x):
        if self.act== "sig":
            return 1/(1+np.exp(-x))
        if self.act== "tanh":
            return np.tanh(x)
        if self.act== "relu":
            return np.maximum(0,x)
        
        
    def act_d(self,x):
        if self.act== "sig":
            s = 1/(1+np.exp(-x))
            return s*(1-s)
        if self.act== "tanh":
            return 1 - np.tanh(x)**2
        if self.act== "relu":
            return (x>0)
        
    def accuracy(y_pred, y):
        count= 0
        for i,j in (y_pred,y):
            if i==j:
                count+= 1
        return count/10
    
    def loss(y_pred, y):
        #  e = (y_pred)*np.log(1 - y_train) + (1 - y_pred)*np.log(y_train)
        e= 0
        for i,j in zip(y_pred, y):
            e+= -((i)*np.log(1- j) + (1- i)*np.log(j))
        return e
```
