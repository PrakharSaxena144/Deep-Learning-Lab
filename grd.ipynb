```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, y):
    return x**2 + y**2 + 10*np.sin(x) + 10*np.cos(y)
```

```python
def gradient_descent(x0, y0, lr=0.05, tol=1e-4, max_iters=1000):
    x, y = float(x0), float(y0)
    path = [(x, y, f(x, y))]

    for i in range(max_iters):
        dfdx = 2*x + 10*np.cos(x)
        dfdy = 2*y - 10*np.sin(y)
        x = x - lr*dfdx
        y = y - lr*dfdy
        path.append((x, y, f(x, y)))
        print(f"Iteration {i}, At point, {x:.3f}, {y:.3f} Loss value: {f(x,y):.4f}")

    print(f"Final point: ({x:.4f}, {y:.4f}), f(x,y) = {f(x,y):.6f}")
    return np.array(path)
```

```python
x_vals = np.linspace(-8, 8, 300)
y_vals = np.linspace(-8, 8, 300)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f(X, Y)
fig = plt.figure(figsize=(11, 8))
ax = fig.add_subplot(111, projection='3d')
ax.set_title("Error Surface")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("f(x,y)")


ax.plot_surface(X, Y, Z, alpha=0.5)
```

```python
plt.figure(figsize=(8,6))
plt.contour(X, Y, Z, levels=40)
lr = 0.01
initial_point = (5,5)

path = gradient_descent(initial_point[0], initial_point[1], lr)
plt.plot(path[:,0], path[:,1],
        marker='o',
        color='blue',
        label=f"lr={lr}")

plt.title("Gradient Descent Trajectories")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
learning_rates = [0.01, 0.05, 0.2]          
initial_points = [(-6,-6), (3,-5), (5,5)]   
steps = 100

fig, axes = plt.subplots(1, len(learning_rates),
                         figsize=(18,5),
                         constrained_layout=True)

for ax, lr in zip(axes, learning_rates):

    contours = ax.contour(X, Y, Z, levels=35)
    ax.clabel(contours, inline=True, fontsize=7)

    for (x0, y0) in initial_points:
        path = gradient_descent(x0, y0, lr=lr)

        ax.plot(path[:,0], path[:,1],
                marker='o', markersize=3, linewidth=1.5,
                label=f"init=({x0},{y0})  f={path[-1,2]:.2f}")

        ax.scatter(path[0,0], path[0,1], s=80, marker='s')
        ax.scatter(path[-1,0], path[-1,1], s=120, marker='*')

    ax.set_title(f"Gradient Descent (lr={lr})")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.grid(True, alpha=0.25)
    ax.legend(fontsize=8)

plt.show()
```

```python
learning_rates = [0.01, 0.25 ,0.2]
initial_points = [(5, 5)]
steps = 100

fig = plt.figure(figsize=(11, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, alpha=0.25)
colors = ['blue', 'green', 'red']

for lr, color in zip(learning_rates, colors):
    for (x0, y0) in initial_points:
        path = gradient_descent(x0, y0, lr=lr)
        xs = path[:, 0]
        ys = path[:, 1]
        zs = path[:, 2]

        # trajectory
        ax.plot(xs, ys, zs,
                linewidth=2,
                color=color,
                alpha=0.9,
                label=f"lr={lr}, init=({x0},{y0})")
        # start point
        ax.scatter(xs[0], ys[0], zs[0],
                   s=60, marker='s', color=color)
        # end point
        ax.scatter(xs[-1], ys[-1], zs[-1],
                   s=90, marker='*', color=color)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x,y)")
ax.set_title("3D Error Surface with GD Trajectories")

handles, labels = ax.get_legend_handles_labels()
unique = dict(zip(labels, handles))
ax.legend(unique.values(), unique.keys(), fontsize=8)

plt.show()
```
